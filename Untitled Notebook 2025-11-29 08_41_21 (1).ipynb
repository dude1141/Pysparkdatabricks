**masking**

```python
%sql

CREATE TABLE ext_catalog.ext_schema.employee
(
    id INT,
    name STRING,
    salary INT
);

INSERT INTO ext_catalog.ext_schema.employee
VALUES
    (1, 'John', 10000),
    (2, 'Mary', 20000),
    (3, 'Mike', 30000);

SELECT * FROM ext_catalog.ext_schema.employee;
```

```python

%sql
CREATE OR REPLACE FUNCTION ext_catalog.ext_schema.masking(salary STRING)
RETURN
    CASE
        WHEN is_account_group_member('Admins') THEN salary
        ELSE '***-**-****'
    END;
```

```python
%sql
select * From ext_catalog.ext_schema.employee;
```

```python
%sql SHOW CATALOGS
```

```python
%sql

ALTER TABLE ext_catalog.ext_schema.employee ALTER COLUMN salary SET MASK ext_catalog.ext_schema.masking;
```
